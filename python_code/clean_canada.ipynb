```python
import pandas as pd
import numpy as np
from string import punctuation as PUNCTUATION
root = "../../Data/"
can_file = root + "canada/psql_extract.csv"
from datetime import datetime
```

```python
hoc_can = pd.read_csv(can_file, sep = "|")
```

```python
hoc_can.sort_values('speechdate', inplace = True)
```

```python
hoc_can['speechdate'] = pd.to_datetime(hoc_can['speechdate'])
hoc_can = hoc_can[hoc_can.speechdate > datetime(1993, 12, 31)]
```

## Remove line breaks, whitespace, commas, punctuation from speech text

```python
drop_char = lambda x, y, z: x.replace(r'\{}'.format(y), z, regex = True, inplace = True)
drop_char(hoc_can, 'n', ' ')
#for i in punctuation:
#    drop_char(hoc_can, i, ' ')
# https://stackoverflow.com/questions/44227748/removing-newlines-from-messy-strings-in-pandas-dataframe-cells
```

```python
hoc_can.to_csv(root + "canada/hoc_can_clean.csv", index = False, sep = "|")
```
